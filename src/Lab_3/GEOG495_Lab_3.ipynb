```python
import pandas as pd
  
# df = pd.read_csv('/content/drive/MyDrive/image_process.csv')
df = pd.read_csv('image_process.csv')

print(df)
```

```python
import altair as alt
import numpy as np

scale2 = alt.Scale(domain=np.array(df['color']),
                  range=np.array(df['color']))

alt.Chart(df).mark_bar().encode(
    x = 'hoursminutes(date):O',
    y = 'percentage', 
    color = alt.Color('color:N', scale=scale2)
).properties(width = 500, height = 500)
```
